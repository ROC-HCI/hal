## 3 Sample

````python
def vary_int(x, percent=5):
    std_dev = percent / 200  # so 95% of values fall within ±percent%
    variation = random.gauss(0, std_dev)
    return int(x * (1 + variation))
````

````python
from re import sub
import random

essential_keys = ['age', 'sex', 'race']

def oversample(d):
    for key, value in d.items():
        if value == "Not applicable":
            del d[key]
    age = d['age']
    non_essential_keys = list(set(list(d.keys())) - set(essential_keys))
    random.shuffle(non_essential_keys)
    total_len = len(non_essential_keys)
    split = total_len//3
    keys1, keys2, keys3 = non_essential_keys[:-split], non_essential_keys[-split:]+non_essential_keys[:split], non_essential_keys[split:]
    random.shuffle(keys1)
    random.shuffle(keys2)
    random.shuffle(keys3)
    keys1 = essential_keys + keys1
    keys2 = essential_keys + keys2
    keys3 = essential_keys + keys3
    sub_dict1 = {k: d[k] for k in keys1 if k in d}
    sub_dict2 = {k: d[k] for k in keys2 if k in d}
    sub_dict3 = {k: d[k] for k in keys3 if k in d}
    sub_dict1['age'] = vary_int(age, 10)
    sub_dict2['age'] = vary_int(age, 10)
    sub_dict3['age'] = vary_int(age, 10)

    return sub_dict1, sub_dict2, sub_dict3

samples = oversample(data[455])
samples
````

## N Sample

````python
import random

def remove_na(obj, token="Not Applicable", case_insensitive=True):
    """Recursively drop keys whose value == token."""
    def is_na(v):
        return isinstance(v, str) and (
            v.casefold() == token.casefold() if case_insensitive else v == token
        )
    if isinstance(obj, dict):
        return {k: remove_na(v, token, case_insensitive) for k, v in obj.items() if not is_na(v)}
    if isinstance(obj, list):
        return [remove_na(x, token, case_insensitive) for x in obj]
    return obj

def _vary_value(x, percent=10, mode="uniform"):
    """Return x varied by ±percent% (uniform or normal)."""
    if mode == "normal":
        std = percent / 200.0  # ~95% within ±percent%
        delta = random.gauss(0, std)
    else:
        delta = random.uniform(-percent / 100.0, percent / 100.0)
    return x * (1 + delta)

def oversample(
    d,
    n=3,
    essential_keys=("age", "sex"),
    jitter_age_percent=10,
    jitter_mode="uniform",
    drop_na=True,
    na_token="Not Applicable",
    case_insensitive=True,
    shuffle_seed=None,
):
    """
    Split dict into n samples:
      - Each sample keeps essential_keys.
      - Non-essential keys are distributed (round-robin) after shuffling.
      - 'age' is jittered per sample if present and numeric.
      - Input dict is not mutated.
    """
    base = remove_na(d, na_token, case_insensitive) if drop_na else dict(d)

    non_essential = [k for k in base.keys() if k not in essential_keys]
    rng = random.Random(shuffle_seed)
    rng.shuffle(non_essential)

    total_len = len(non_essential)
    ## Each sample must have at least 4 unique characteristics
    if total_len < n*4:
      n = total_len//4
    split_size = total_len//(n+1)
    expanded_temp_list = non_essential*3

    buckets = []
    for i in range(total_len, 2*total_len, split_size):
      bucket = expanded_temp_list[i-split_size: i+split_size]
      random.shuffle(bucket)
      buckets.append(bucket)

    samples = []
    for i in range(n):
        keys_i = list(essential_keys) + buckets[i]
        sub = {k: base[k] for k in keys_i if k in base}
        if "age" in sub and isinstance(sub["age"], (int, float)):
            sub["age"] = int(round(_vary_value(sub["age"], jitter_age_percent, jitter_mode)))
        samples.append(sub)
    return samples
````

## Oversample List

````python
def oversample_list(personas):
  oversampled_personas = []
  for idx, persona in enumerate(personas):
    try:
      oversampled = oversample(persona, n=4, jitter_age_percent=10)
      for i, item in enumerate(oversampled):
        item['parent_id'] = idx
        item['id'] = idx*4 + i
      oversampled_personas.extend(oversampled)
    except:
      print("Error in persona:", idx)
  return oversampled_personas
````

````python
import os
import getpass
from openai import OpenAI

def _set_env(var: str):
    if var not in os.environ:
        os.environ[var] = getpass.getpass(f"{var}: ")

_set_env("OPENAI_API_KEY")
````

````python
# @title
class TinyAgent:

    def __init__(self, model, tokenizer=None, debug=False):

        self.model = model
        self.messages = []
        self.max_tokens = 3072
        self.debug = False
        self.reasoning_effort = "low"
        self.temperature = 1

    def clear_messages(self):
        self.messages = list()

    def add_message(self,message_type, message):
        self.messages.append({"role": message_type, "content":message})

    def add_system_message(self, message):
        self.add_message("system", message)

    def add_user_message(self, message):
        self.add_message("user", message)

    def add_assistant_message(self, message):
        self.add_message("assistant", message)

    def set_max_tokens(self, max_tokens):
        self.max_tokens = max_tokens

    def set_debug(self, debug):
        self.debug = debug

    def set_reasoning_effort(self, reasoning_effort):
        self.reasoning_effort = reasoning_effort

    def call(self, prompt="", response_type="text", cache=True):
        messages = self.messages.copy()
        if prompt:
            messages.append({"role": "user", "content":prompt})
        if cache:
            self.add_user_message(prompt)

        if "gpt-5" in self.model:
            response = client.responses.create(
                model=self.model,
                input=messages,
                reasoning={"effort": self.reasoning_effort},
                text={
                    "format": {
                      "type": response_type
                    },
                    "verbosity": "low"
                  },
            )
            reply = response.output_text

        elif "gpt-4" in self.model:
            response = client.responses.create(
              model=self.model,
              input=messages,
              temperature=self.temperature,
              max_output_tokens=self.max_tokens,
              top_p=1,
              text={
                "format": {
                  "type": response_type # "text", "json_object"
                }
              }
            )
            reply = response.output_text
        if self.debug:
            print(reply)
        if cache:
            self.add_assistant_message(reply)
        return reply

    def load_json(self,s):
        import json
        try:
            return json.loads(s)
        except json.JSONDecodeError:
            return None

    def call_json(self, prompt=""):
        self.add_system_message("Reply must be JSON format.")
        reply = self.call(prompt=prompt, response_type="json_object")
        if not reply:
            print("Empty reply")
            return None

        reply = reply.strip()
        if reply.startswith("```json"):
            reply = reply[len("```json"):].strip()
            if reply.endswith("```"):
                reply = reply[:-3].strip()

        # Use OR, and guard length
        if not (reply.startswith("{") and reply.endswith("}")):
            print("Not JSON structure")
            return None

        try:
            return self.load_json(reply)
        except Exception:
            print("Error parsing JSON")
            return None
````

````python
negative_personality_traits = [
    # Interpersonal Traits
    "Arrogant", "Narcissistic", "Manipulative", "Machiavellian",
    "Hostile", "Aggressive", "Jealous", "Possessive",
    "Dominating", "Deceitful",

    # Emotional Traits
    "Anxious", "Moody", "Irritable", "Impulsive",
    "Resentful", "Bitter", "Overly Sensitive",

    # Cognitive / Attitudinal Traits
    "Rigid", "Dogmatic", "Cynical", "Pessimistic", "Paranoid",

    # Motivational / Behavioral Traits
    "Lazy", "Unmotivated", "Disorganized", "Undisciplined",
    "Reckless", "Procrastinating"
]
````

````python
import json

def _persona_generation_prompt(persona_json, neg_trait=""):
  additional_prompt = ""
  if neg_trait:
    additional_prompt = "Give the person this negative trait: " + neg_trait

  return """
From the given JSON file for a persona, come up with the following information about this hypothetical person. Make up realistic information with vivid detail without deviating from the given information.
Our goal is to create a realistic description of a real person, not an idealistic or perfect persona.
""" + additional_prompt + """

## Required Fields:
- firstname
- lastname
- interests
- values
- biography
- medical condition
- reason for a medical visit

## Instructions:
- Interests: Select specific, realistic topics or activities the individual enjoys (e.g., politics, family, hobbies, job), based on provided data. Ensure choices are grounded and plausible.
- Values: Identify and list key values grounded in the persona's traits or narrative.
- Biography: Write a thorough, detailed, and specific narrative integrating interests, values, worries, happiness sources, strengths, weaknesses, and other contextual elements. Maintain authenticity and nuance without idealization. Make it human, not perfect.
- medical condition: Give them a specific medical condition they have been suffering from
- reason for medical visit: The person is attending a medical visit to a doctor or healthcare provider. It could be first time, or recurrent visit. Specify their reason for visit today.

If any required data is missing from the JSON file, infer the most plausible value from available data. If reasonable inference is not possible, use the string "unknown" for that field. For array fields ('interests', 'values'), supply an empty array if no reasonable inference can be made. Strictly adhere to the output schema, regardless of data completeness or clarity.
After generating the output, validate that the JSON structure exactly matches the required schema and all fields are filled as specified.

## Output Format
Return a single JSON object with this structure:
{
"firstname": string,
"lastname": string,
"interests": [array of strings],
"values": [array of strings],
"biography": string (up to 1000 tokens),
"medical_condition": string,
"reason_for_visit": string
}

Input data:
""" + json.dumps(persona_json, indent=2)
````

````python
import random, datetime


def generate_full_persona(persona):
  model="gpt-4.1"
  try:
    persona_json = persona.copy()
    del persona_json['parent_id']
    del persona_json['id']
    random.seed(datetime.datetime.now().timestamp())
    probability = 0.1
    neg_trait = ""
    if random.random() < probability:
      neg_trait = random.choice(negative_personality_traits)
      persona_prompt = _persona_generation_prompt(persona_json, neg_trait)
      # print(f"Using negative trait: {neg_trait}")
    else:
      persona_prompt = _persona_generation_prompt(persona_json)

    agent = TinyAgent(model)
    agent.set_debug(False)
    agent.clear_messages()

    # print(f"Prompt: {persona_prompt}")
    agent.add_user_message(persona_prompt)
    result = agent.call_json()
    result["neg_trait"] = neg_trait
    # result["parent_id"] = persona["parent_id"]
    # result["id"] = persona["id"]
    # print(f"Result: {result}")
    return persona | result
  except Exception as e:
    print(f"Error generating persona: {e}")
    return None
````

````python
client = OpenAI()

from concurrent.futures import ThreadPoolExecutor, as_completed

def process_fast(max_workers, data, results, print_after=1):
    tasks = data.copy()

    print(f"Total task: {len(tasks)} | Number of Workers: {max_workers}")

    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        futs = [ex.submit(generate_full_persona, row) for row in tasks]

        for i, fut in enumerate(as_completed(futs), 1):
            r = fut.result()
            if r:
                results.append(r)
                if i % print_after == 0:
                  print(f"[Saved persona {i}] Progress {i}/{len(futs)} | Current total saved {len(results)}")
            else:
              print(f"[Skipped persona {i}] Progress {i}/{len(futs)} | Current total saved {len(results)}")

    print(f"Done. Saved {len(results)}")
````

````python
llms = ['gpt-4.1', 'gpt-4.1-mini', 'gpt-4.1-nano', 'gpt-5', 'gpt-5-mini', 'gpt-5-nano']
reasoning_efforts = ['high', 'medium', 'low']
````

````python
def _dialogue_generation_prompt(persona):
  biography = persona['biography']
  medical_condition = persona['medical_condition']
  reason_for_visit = persona['reason_for_visit']
  return """Generate a full conversation of the following person at the doctor's visit.
Give them a unique personality based on their biography. Follow a linguistic style suitable for the person. Vary the statement lengths to make it natural.
Don't say anything else.

Dialogue format:
---
P: ...
D: ...

Input data: \nBiography: """ + biography + "\nMedical condition: " + medical_condition + "\nReason's for visit: " + reason_for_visit + "\n"
````

````python
print(_dialogue_generation_prompt(synthlabs_test_generated[0]))
````

````python
import random, datetime

def generate_dialogue(persona, generation_num):
  try:
    persona_json = persona.copy()
    random.seed(datetime.datetime.now().timestamp())

    persona_prompt = _dialogue_generation_prompt(persona_json)

    dialogues = []
    for _ in range(generation_num):
      model = random.choice(llms)
      try:
        reasoning_effort = None
        agent = TinyAgent(model)
        if model.startswith("gpt-5"):
          reasoning_effort = random.choice(reasoning_efforts)
          agent.set_reasoning_effort(reasoning_effort)
        agent.set_debug(False)
        agent.clear_messages()
        agent.add_user_message(persona_prompt)
        result = agent.call()
        dialogues.append({
          "model" : model,
          "reasoning_effort": reasoning_effort,
          "dialogue" : result,
          "score": 0
        })
      except:
        print(f"Error generating dialogue for persona {persona['id']} with model {model}")
        return None

    return persona | {"dialogues": dialogues}
  except Exception as e:
    print(f"Error generating persona: {e}")
    return None
````

````python
def clean_dialogue(text):
    if pd.isna(text):
        return text
    # Replace multiple newlines with one
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'---', '', text)
    # Remove everything inside first parentheses (non-nested)
    text = re.sub(r'\([^)]*\)', '', text)
    text = re.sub(r'[ \t]+', ' ', text)
    text += '\n</dialogue>\n'
    return text.strip()
````

````python
import pandas as pd
import random
import re

NUM_CHOSEN_DIALOGUE = 4

def dialogue_to_csv(data, csvname):
    dialogue_dict = []
    prompt_dict = []
    for item in data:
        parent_id = item["parent_id"]
        persona_id = item["id"]
        prompt_dict.append({"persona_id": persona_id, "prompt": make_prompt(item)})
        chosen_dialogues = random.sample(item["dialogues"], NUM_CHOSEN_DIALOGUE)
        for did, dialogues in enumerate(chosen_dialogues):
            dialogue_id = persona_id * NUM_CHOSEN_DIALOGUE + did
            dialogues["parent_id"] = parent_id
            dialogues["persona_id"] = persona_id
            dialogues["dialogue_id"] = dialogue_id
            dialogues["dialogue"] = clean_dialogue(dialogues["dialogue"])
            dialogue_dict.append(dialogues)

    # Convert to DataFrame and save as CSV
    df = pd.DataFrame(dialogue_dict)
    prompt_df = pd.DataFrame(prompt_dict)
    df.to_csv(csvname, index=False)
    prompt_df.to_csv("prompts_"+csvname, index=False)
    print(f"✅ Saved {len(df)} dialogues to {csvname}")


dialogue_to_csv(synthlabs_train_dialogues, "synthlabs_dialogues_train.csv")
dialogue_to_csv(synthlabs_test_dialogues, "synthlabs_dialogues_test.csv")
````

````python
import numpy as np
import pandas as pd

total_n = df_train["hal16_score"].notna().sum()

summary = []
for model, g in df_train.groupby("model"):
    x = g["hal16_score"].dropna().values
    n = len(x)
    mean = np.mean(x)
    std = np.std(x, ddof=1)
    ci95 = 1.96 * std / np.sqrt(n)
    freq_pct = 100.0 * n / total_n

    summary.append({
        "model": model,
        "mean_hal16": round(mean, 2),
        "ci95_low": round(mean - ci95, 2),
        "ci95_high": round(mean + ci95, 2),
        "n": n,
        "frequency_pct": round(freq_pct, 2),
    })

summary_df = pd.DataFrame(summary).sort_values("mean_hal16", ascending=False)
summary_df
````
